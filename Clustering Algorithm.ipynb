## Clustering Algorithm

```python
import pandas as pd
from sklearn.datasets import load_iris

#Load thr iris dataset
iris=load_iris()
iris_data=pd.DataFrame(data=iris.data,columns=iris.feature_names)

print(iris_data.head())
```

### implementing KMeans Clustering

#### KMeans Clustering 
Description of KMeans Clustering: KMeans clustering is an unsupervised learning algorithm that partitions the dataset into K distinct clusters. It works by initializing K centroids randomly, assigning each data point to the nearest centroid, and then recalculating the centroids based on the assigned points. This process is repeated until the centroids no longer change significantly.

Suitability for the Iris Dataset: KMeans is suitable for the Iris dataset because it has well-defined clusters based on the features (sepal length, sepal width, petal length, petal width). The dataset contains three species of iris flowers, which can be effectively grouped into clusters.

```python

#KMeans Clustering 
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
#Appying  Clustering
Kmeans=KMeans(n_clusters=3,random_state=42)
iris_data['Clusters']=Kmeans.fit_predict(iris_data)

#visualize the clusters
plt.figure(figsize=(10,6))
plt.scatter(iris_data['sepal length (cm)'], iris_data['sepal width (cm)'], c=iris_data['Clusters'], cmap='viridis')
plt.title('KMeans Clustering for Iris Dataset')
plt.xlabel('sepal length (cm)')
plt.ylabel('sepal width (cm)')
plt.show()
```

## Implementing Hierarchical Clustering

```python
#Hierarchical Clustering
from sklearn.cluster import AgglomerativeClustering
import seaborn as sns
import scipy.cluster.hierarchy as sch

#Applying Hierarchical Clustering
hierarchical=AgglomerativeClustering(n_clusters=3)
iris_data['Hierarchical_Cluster']=hierarchical.fit_predict(iris_data)

#visualizing using dendrogram
plt.figure(figsize=(10,7))
dendrogram = sch.dendrogram(sch.linkage(iris_data, method='ward'))
plt.title('Dendrogram for Hierarchical Clustering')
plt.xlabel=('Samples')
plt.ylabel=('Distance')
plt.show()

#visualize the clusters
plt.figure(figsize=(10, 6))
sns.scatterplot(x='sepal length (cm)', y='sepal width (cm)', hue='Hierarchical_Cluster', data=iris_data, palette='viridis')
plt.title('Hierarchical Clustering of Iris Dataset')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.show()


```
